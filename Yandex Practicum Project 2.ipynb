## Предобработка данных

```python
# подсчёт доли пропусков
df.isna().mean()
```

В основном `NaN` значение обнаружено в столбцах с пропущенными значениями. В столбцах `days_employed` и `total_income` доля пропусков составляет 10%. Также, пропуски для этих столбцов является одинаковыми, то есть если значение в столбце `days_employed` является пустым то соответственно значение в столбце `total_income` является тоже пустым, и наоборот. Возможные причины появления пропусков в данных могут быть:
* клиент является пенсионером то есть не имеет дохода только пенсия
* клиент не имеет какое то точное число заработка
* клиент может быть безработным

```python
# Замена пропусков с медианным значением по столбцу days_employed
days_employed_median = df['days_employed'].median()
df['days_employed'].fillna(days_employed_median, inplace=True)
```

```python
# Замена пропусков с медианным значением по столбцу total_income
total_income_median = df['total_income'].median()
df['total_income'].fillna(total_income_median, inplace=True)
```

Замена пропусков с медианным значением по столбцу является очень распрастраненным методом для количественных переменных.

```python
# проверочный подсчёт доли пропусков
df.isna().mean()
```

### Категоризация дохода.

Создаём функию которая будет присваивать определенную категорию в столбец `total_income_category` относительно дохода клиента, то есть столбца `total_income`.

На основании диапазонов, указанных ниже, создаём столбец `total_income_category` с категориями:
   - 0–30000 — 'E';
   - 30001–50000 — 'D';
   - 50001–200000 — 'C';
   - 200001–1000000 — 'B';
   - 1000001 и выше — 'A'.
   
Например, кредитополучателю с доходом 25000 нужно назначаем категорию 'E', а клиенту, получающему 235000, — 'B'.

```python
def income_category(income):
    if income <= 30000:
        return 'E'
    elif 30001 <= income <= 50000:
        return 'D'
    elif 50001 <= income <= 200000:
        return 'C'
    elif 200001 <= income <= 1000000:
        return 'B'
    elif income >= 1000001:
        return 'A'
    
df['total_income_category'] = df['total_income'].apply(income_category)
```

### Категоризация целей кредита.

Создаём функцию, которая на основании данных из столбца purpose сформирует новый столбец purpose_category, в который войдут следующие категории:
   - 'операции с автомобилем',
   - 'операции с недвижимостью',
   - 'проведение свадьбы',
   - 'получение образования'.
    
Например, если в столбце purpose находится подстрока 'на покупку автомобиля', то в столбце purpose_category появится строка 'операции с автомобилем'.

```python
def purpose_category(purpose):
    if 'автомобил' in purpose:
        return 'операции с автомобилем'
    elif 'недвижимост' in purpose or 'жиль' in purpose:
        return 'операции с недвижимостью'
    elif 'свадьб' in purpose:
        return 'проведение свадьбы'
    elif 'образован' in purpose:
        return 'получение образования'
    
df['purpose_category'] = df['purpose'].apply(purpose_category)
```
